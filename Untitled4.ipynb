```python
# cleaning the data
import pandas as pd
#Step 1: Load and combine original Excel sheets
grassland = pd.concat(pd.read_excel("/content/drive/MyDrive/Copy of Bird_Monitoring_Data_GRASSLAND.XLSX.xlsx", sheet_name=None), ignore_index=True)
forest = pd.concat(pd.read_excel("/content/drive/MyDrive/bird_forest", sheet_name=None), ignore_index=True)

# Step 2: Add habitat labels
grassland["Habitat"] = "Grassland"
forest["Habitat"] = "Forest"

# Step 3: Combine datasets
combined_df = pd.concat([grassland, forest], ignore_index=True)

# Step 4: Standardize column names
combined_df.columns = combined_df.columns.str.strip().str.lower().str.replace(' ', '_')

# Step 5: Rename common fields
combined_df = combined_df.rename(columns={
    'common_name': 'species',
    'initial_three_min_cnt': 'count'
})

# Step 6: Select relevant columns
relevant_cols = ['date', 'start_time', 'end_time', 'species', 'count',
                 'temperature', 'humidity', 'sky', 'wind', 'observer', 'habitat']
filtered_df = combined_df[relevant_cols].copy()

# Step 7: Drop rows missing essential fields
filtered_df = filtered_df.dropna(subset=['species', 'count', 'date'])

# Step 8: Convert 'count' to numeric and drop invalid entries
filtered_df['count'] = pd.to_numeric(filtered_df['count'], errors='coerce')
filtered_df = filtered_df.dropna(subset=['count'])
filtered_df['count'] = filtered_df['count'].astype(int)

# Step 9: Convert 'date' to datetime
filtered_df['date'] = pd.to_datetime(filtered_df['date'], errors='coerce')
filtered_df = filtered_df.dropna(subset=['date'])

# Step 10: Handle missing time values
filtered_df['start_time'] = filtered_df['start_time'].fillna('00:00')
filtered_df['end_time'] = filtered_df['end_time'].fillna('00:00')

# Step 11: Fill missing environment & observer data with 'Unknown'
for col in ['temperature', 'humidity', 'sky', 'wind', 'observer']:
    if col in filtered_df.columns:
        filtered_df[col] = filtered_df[col].fillna('Unknown')

# Final output
print("Final cleaned data :", filtered_df.shape)
print(filtered_df.head())
```

```python
#  performing analysis on the cleaned data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Extract temporal features
filtered_df['year'] = filtered_df['date'].dt.year
filtered_df['month'] = filtered_df['date'].dt.month

# Derive seasons from month
filtered_df['season'] = filtered_df['month'] % 12 // 3 + 1
season_map = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'}
filtered_df['season'] = filtered_df['season'].map(season_map)

# Convert start and end times to hour of day
filtered_df['start_hour'] = pd.to_datetime(filtered_df['start_time'], errors='coerce').dt.hour
filtered_df['end_hour'] = pd.to_datetime(filtered_df['end_time'], errors='coerce').dt.hour

# Seasonal Trend: Total bird count per season per year
seasonal_trends = filtered_df.groupby(['year', 'season'])['count'].sum().reset_index()

# Time of Day Trend: Average count per start_hour
time_trends = filtered_df.groupby('start_hour')['count'].mean().reset_index()

# Plot 1: Seasonal Trends
plt.figure(figsize=(10, 5))
sns.lineplot(data=seasonal_trends, x='year', y='count', hue='season', marker='o')
plt.title("Bird Sightings by Season and Year")
plt.ylabel("Total Bird Count")
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Observation Time Trends
plt.figure(figsize=(8, 4))
sns.lineplot(data=time_trends, x='start_hour', y='count', marker='o')
plt.title("Average Bird Count by Start Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Average Bird Count")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Spatial Analysis: Group by habitat (location type) and plot name

# 1. Total unique species and bird counts by habitat
habitat_summary = filtered_df.groupby('habitat').agg(
    total_observations=('count', 'sum'),
    unique_species=('species', 'nunique'),
    total_records=('species', 'count')
).reset_index()

# 2. Plot-level analysis: total observations and species richness per plot
if 'plot_name' in filtered_df.columns:
    plot_summary = filtered_df.groupby('plot_name').agg(
        total_observations=('count', 'sum'),
        unique_species=('species', 'nunique'),
        total_records=('species', 'count')
    ).reset_index().sort_values(by='unique_species', ascending=False)
else:
    plot_summary = pd.DataFrame(columns=['plot_name', 'total_observations', 'unique_species', 'total_records'])

# Plot: Habitat-level species highhness and observation volume
plt.figure(figsize=(8, 5))
sns.barplot(data=habitat_summary, x='habitat', y='unique_species')
plt.title('Unique Species Observed by Habitat')
plt.ylabel('Number of Unique Species')
plt.xlabel('Habitat Type')
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# Plot:  plots by unique species richness
if not plot_summary.empty:
    top_plots = plot_summary.head(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(data=top_plots, y='plot_name', x='unique_species', palette='viridis')
    plt.title('Top 10 Plots by Unique Species Count')
    plt.xlabel('Unique Species')
    plt.ylabel('Plot Name')
    plt.tight_layout()
    plt.show()

habitat_summary, plot_summary.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Making a copy to work with cleaned data
env_df = filtered_df.copy()

# Convert temperature and humidity to numeric
env_df['temperature'] = pd.to_numeric(env_df['temperature'], errors='coerce')
env_df['humidity'] = pd.to_numeric(env_df['humidity'], errors='coerce')

# 1. Temperature vs Bird Count
plt.figure(figsize=(8, 5))
sns.scatterplot(data=env_df, x='temperature', y='count', hue='habitat')
plt.title("Bird Count vs Temperature")
plt.xlabel("Temperature (C)")
plt.ylabel("Bird Count")
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Humidity vs Bird Count
plt.figure(figsize=(8, 5))
sns.scatterplot(data=env_df, x='humidity', y='count', hue='habitat')
plt.title("Bird Count vs Humidity")
plt.xlabel("Humidity (%)")
plt.ylabel("Bird Count")
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Sky Condition impact
plt.figure(figsize=(8, 5))
sns.boxplot(data=env_df, x='sky', y='count', hue='habitat')
plt.title("Bird Count by Sky Condition")
plt.xlabel("Sky")
plt.ylabel("Bird Count")
plt.grid(True)
plt.tight_layout()
plt.show()

# 4. Wind Condition impact
plt.figure(figsize=(8, 5))
sns.boxplot(data=env_df, x='wind', y='count', hue='habitat')
plt.title("Bird Count by Wind Condition")
plt.xlabel("Wind")
plt.ylabel("Bird Count")
plt.grid(True)
plt.tight_layout()
plt.show()

# 5. Disturbance effect
if 'disturbance' in env_df.columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(data=env_df, x='disturbance', y='count', hue='habitat')
    plt.title("Bird Count by Disturbance Level")
    plt.xlabel("Disturbance")
    plt.ylabel("Bird Count")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    print("No 'disturbance' column found in dataset.")
```

```python
# Observer Trends Analysis

# Observer Bias: Total observations and species reported per observer
observer_summary = filtered_df.groupby('observer').agg(
    total_observations=('count', 'sum'),
    unique_species=('species', 'nunique'),
    total_records=('species', 'count')
).reset_index().sort_values(by='total_observations', ascending=False)

# Plot 1: Observations per Observer
plt.figure(figsize=(10, 6))
sns.barplot(data=observer_summary.head(10), x='total_observations', y='observer', palette='crest')
plt.title("Top Observers by Total Bird Observations")
plt.xlabel("Total Observations")
plt.ylabel("Observer")
plt.tight_layout()
plt.show()

# Visit Patternss
if 'visit' in filtered_df.columns:
    visit_summary = filtered_df.groupby('visit').agg(
        total_observations=('count', 'sum'),
        unique_species=('species', 'nunique'),
        total_records=('species', 'count')
    ).reset_index().sort_values(by='visit')

    # Plot 2: Observations by Visit Number
    plt.figure(figsize=(8, 5))
    sns.lineplot(data=visit_summary, x='visit', y='unique_species', marker='o')
    plt.title("Species Richness by Visit Number")
    plt.xlabel("Visit")
    plt.ylabel("Unique Species Count")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
else:
    visit_summary = pd.DataFrame(columns=['visit', 'total_observations', 'unique_species', 'total_records'])

observer_summary.head(), visit_summary.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Making a copy of cleaned data
df = filtered_df.copy()

# 1. Convert 'distance' to numeric (if present)
if 'distance' in df.columns:
    df['distance'] = pd.to_numeric(df['distance'], errors='coerce')

    # Plot: Average distance by species (top 10 most observed)
    top_species = df['species'].value_counts().head(10).index
    distance_by_species = df[df['species'].isin(top_species)].groupby('species')['distance'].mean().sort_values()

    plt.figure(figsize=(10, 6))
    sns.barplot(x=distance_by_species.values, y=distance_by_species.index, palette="mako")
    plt.title("Average Distance of Top 10 Species Observed")
    plt.xlabel("Average Distance (meters)")
    plt.ylabel("Species")
    plt.tight_layout()
    plt.show()
else:
    print("No 'distance' column found in dataset.")

# 2. Flyover Frequency
if 'flyover_observed' in df.columns:
    # Clean flyover values
    df['flyover_observed'] = df['flyover_observed'].astype(str).str.lower().str.strip()

    flyover_summary = df.groupby('flyover_observed')['count'].sum().reset_index().sort_values(by='count', ascending=False)

    plt.figure(figsize=(6, 4))
    sns.barplot(data=flyover_summary, x='flyover_observed', y='count', palette='Set2')
    plt.title("Total Bird Count by Flyover Observation")
    plt.xlabel("Flyover Observed")
    plt.ylabel("Total Bird Count")
    plt.tight_layout()
    plt.show()
else:
    print("No 'flyover_observed' column found in dataset.")
```

```python
# Species Analysis
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Species Diversity per Habitat
diversity_by_habitat = filtered_df.groupby('habitat')['species'].nunique().reset_index(name='unique_species')

plt.figure(figsize=(8, 5))
sns.barplot(data=diversity_by_habitat, x='habitat', y='unique_species', palette='Set2')
plt.title("Unique Species Observed per Habitat")
plt.xlabel("Habitat Type")
plt.ylabel("Number of Unique Species")
plt.grid(axis='y')
plt.tight_layout()
plt.show()

# 2. Most Frequently Observed Species
top_species = (
    filtered_df.groupby('species')['count'].sum()
    .reset_index()
    .sort_values(by='count', ascending=False)
    .head(10)
)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_species, x='count', y='species', palette='viridis')
plt.title("Top 10 Most Frequently Observed Species")
plt.xlabel("Total Count")
plt.ylabel("Species")
plt.tight_layout()
plt.show()
```
